```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('data/train.csv')
data.head(30)
```

```python
data.info()
```

```python
data.describe()
```

```python
output = pd.DataFrame(columns = ['id', 'Personality'])
```

```python
output['id'] = data['id']
output['Personality'] = 'NaN'
output.head()
```

```python
output['Personality'] = data.apply(lambda x: 'Introvert' if x['Stage_fear'] == 'Yes' or x['Drained_after_socializing'] == 'Yes' else 'Extrovert', axis=1)
```

```python
output.head(10)
```

```python
accuracy = (data['Personality'] == output['Personality']).mean()
print(f"Accuracy: {accuracy * 100:.2f}%")
```
